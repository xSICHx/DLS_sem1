```python
no_internet_cols = []
no_phone_cols = []
for cat in cat_cols:
    if "object" == df[cat].dtypes and 'No internet service' in df[cat].values:
        no_internet_cols.append(cat)
    if "object" == df[cat].dtypes and 'No phone service' in df[cat].values:
        no_phone_cols.append(cat)

print(f"Columns with category \"No internet service\":\n{no_internet_cols}\n")
print(f"Columns with category \"No phone service\":\n{no_phone_cols}")
```

```python
for col in no_internet_cols:
    df[col] = df[col].replace("No internet service", "No")
for col in no_phone_cols:
    df[col] = df[col].replace("No phone service", "No")
```

Оформим это преобразование функцией

```python
def simplify_dataset(data):
    df = data.copy()
    no_internet_cols = []
    no_phone_cols = []
    for cat in cat_cols:
        if "object" == df[cat].dtypes and 'No internet service' in df[cat].values:
            no_internet_cols.append(cat)
        if "object" == df[cat].dtypes and 'No phone service' in df[cat].values:
            no_phone_cols.append(cat)

    for col in no_internet_cols:
        df[col] = df[col].replace("No internet service", "No")
    for col in no_phone_cols:
        df[col] = df[col].replace("No phone service", "No")
    return df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = simplify_dataset(df)

colors = ['skyblue', 'salmon', 'lightgreen', 'gold', 'lightcoral', 'violet', 'cyan', 'orange',
          'pink', 'lime', 'teal', 'magenta', 'brown', 'gray', 'olive', 'purple']

fig, axes = plt.subplots(4, 4, figsize=(16, 16)) 
axes_flat = axes.flatten()

for i, col in enumerate(cat_cols):
    value_counts = df[col].value_counts()
    value_counts.plot(kind='bar', ax=axes_flat[i], color=colors[i % len(colors)])
    axes_flat[i].set_title(f'Распределение {col}')
    axes_flat[i].set_xlabel(col)
    axes_flat[i].set_ylabel('Количество')
    axes_flat[i].tick_params(axis='x', rotation=45)


plt.suptitle('Bar-диаграммы категориальных признаков после преобразования', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder, OrdinalEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

Сделаем сразу пайплайном. Плюс посмотрим, как подействует робастное масштабирование (RobustScaler)

Кроме того, сделаем свой трансформер для того, чтобы заполнить значения TotalSpent медианным значением **только** из тренировочной выборки и закинуть это в пайплайн

```python
df = pd.read_csv('./train.csv')
df = simplify_dataset(df)

df["TotalSpent"] = pd.to_numeric(df['TotalSpent'], errors='coerce').fillna(1)
df = df.drop(index = df[ (df["TotalSpent"] > 1000) & (df["ClientPeriod"] < 2) ].index)

# print(df.shape)


preprocessor = make_column_transformer(
        (RobustScaler(), num_cols),
        (OneHotEncoder(drop='first'), cat_cols) 
)

# print(
#     preprocessor.fit_transform(df.drop("Churn", axis=1)).shape
#       )


pipeline = make_pipeline( 
    preprocessor, 
    LogisticRegression(max_iter=1000) 
)


param_grid = { 
    'logisticregression__C': [10, 1, 0.1, 0.01, 0.001],  
    'logisticregression__penalty':['l2'],
    'logisticregression__class_weight': ['balanced']
}


grid_search_robust_lg = GridSearchCV( 
    estimator=pipeline, 
    param_grid=param_grid, 
    scoring='roc_auc',  
    cv=10,         
    refit=True, 
    n_jobs=-1, 
    verbose=1 
)


grid_search_robust_lg.fit(df.drop("Churn", axis=1), df["Churn"])
```

```python
print("Лучшие параметры:", grid_search_robust_lg.best_params_)
print("Roc-auc на тестовых данных:", grid_search_robust_lg.best_score_)
```

```python
import catboost 


df = pd.read_csv('./train.csv')
df = simplify_dataset(df)
df["TotalSpent"] = pd.to_numeric(df['TotalSpent'], errors='coerce').fillna(1)


boosting_model = catboost.CatBoostClassifier(
        cat_features=cat_cols,
        logging_level="Silent",
        eval_metric="AUC:hints=skip_train~false",
        auto_class_weights='Balanced',
        random_seed=0,
        grow_policy='Lossguide'
)

PARAMS = {
    "learning_rate": [0.03], #[0.03, 0.05],
    "l2_leaf_reg": [5],      #[3, 5, 10, 12],
    "n_estimators": [150]    #[150, 200]
}


boosting_model.grid_search(PARAMS,
                           df.drop("Churn", axis=1), df["Churn"],
                           cv = 5,
                        #    plot=True,
                           refit=True, verbose=False);
```

```python
print("Лучшие параметры:", boosting_model.get_params())
print("Точность на валидационных данных:", boosting_model.best_score_)
```

```python
submission = pd.read_csv('./submission.csv')

X_test = pd.read_csv('./test.csv')
X_test = simplify_dataset(X_test)
X_test["TotalSpent"] = pd.to_numeric(X_test['TotalSpent'], errors='coerce').fillna(1)


best_model = boosting_model

submission['Churn'] = best_model.predict_proba(X_test)[:, 1]
print(submission)
submission.to_csv('./cat_submission.csv', index=False)
```

## Случайный лес

```python
from sklearn.ensemble import RandomForestClassifier



df = pd.read_csv('./train.csv')
df = simplify_dataset(df)
df["TotalSpent"] = pd.to_numeric(df['TotalSpent'], errors='coerce').fillna(1)


preprocessor = make_column_transformer(
        (OneHotEncoder(drop='first'), cat_cols)
)

pipeline = make_pipeline( 
    preprocessor, 
    RandomForestClassifier(random_state=42)
)



param_grid = { 
    "randomforestclassifier__n_estimators": [150, 200, 300], 
    "randomforestclassifier__class_weight": ['balanced', 'balanced_subsample'],
    "randomforestclassifier__max_depth": [2, 4, 6, 8]
}


gd_rf = GridSearchCV( 
    estimator=pipeline, 
    param_grid=param_grid, 
    scoring='roc_auc',  
    cv=10,         
    refit=True, 
    n_jobs=-1, 
    verbose=1 
)


gd_rf.fit(df.drop("Churn", axis=1), df["Churn"])
```

Выпишите, какое лучшее качество и с какими параметрами вам удалось получить



```python
print("Лучшие параметры:", gd_rf.best_params_)
print("Точность на валидационных данных:", gd_rf.best_score_)
```

```python
X_test = pd.read_csv('./test.csv')
X_test = simplify_dataset(X_test)
submission = pd.read_csv('./submission.csv')
X_test["TotalSpent"] = pd.to_numeric(X_test['TotalSpent'], errors='coerce').fillna(1)


best_model = gd_rf



submission['Churn'] = best_model.predict_proba(X_test)[:, 1]
print(submission)
submission.to_csv('./rf_submission.csv', index=False) 
```
